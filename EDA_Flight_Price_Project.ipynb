# Feature Enginnering

```python
df.head(2)
```

```python
### 📅 Extracting Features from Date of Journey

df["Date"]=df["Date_of_Journey"].str.split("/").str[0]
df["Month"]=df["Date_of_Journey"].str.split("/").str[1]
df["Year"]=df["Date_of_Journey"].str.split("/").str[2]
```

```python
df.head(2)
```

```python
df.info()
```

After extracting day, month, and year from the Date_of_Journey feature, these columns are in object data type and need to be converted into integer format for further analysis.

```python
# Typecasting

df["Date"]=df["Date"].astype(int)
df["Month"]=df["Month"].astype(int)
df["Year"]=df["Year"].astype(int)
```

```python
df.info()
```

```python
# Now drop the column Date Of Journey--

df.drop("Date_of_Journey",axis=1, inplace=True)
```

```python
df.head(2)
```

⏰ Feature Engineering on Arrival Time

```python
df["Arrival_Time"]=df["Arrival_Time"].apply(lambda x:x.split(' ')[0])
df["Arrival_Hour"]=df["Arrival_Time"].str.split(":").str[0]
df["Arrival_Minute"]=df["Arrival_Time"].str.split(":").str[1]
```

```python
df.head(2)
```

```python
df.info()
```

After extracting arrival hour and minutes, these features are converted into integer format for further use.

```python
df["Arrival_Hour"]=df["Arrival_Hour"].astype(int)
df["Arrival_Minute"]=df["Arrival_Minute"].astype(int)

                      
```

```python
df.head(2)
```

```python
df.info()
```

```python
df["Dep_Hour"]=df["Dep_Time"].str.split(":").str[0]
df["Dep_Minute"]=df["Dep_Time"].str.split(":").str[1]
```

```python
df["Dep_Hour"]=df["Dep_Hour"].astype(int)
df["Dep_Minute"]=df["Dep_Minute"].astype(int)
```

```python
df["Duration_Hours"]=df["Duration_Hours"].fillna(0).astype(int)
df["Duration_Minutes"]=df["Duration_Minutes"].fillna(0).astype(int)
```

--The 'Total_Stops' feature represents the number of stops between the source and destination and is an important factor affecting flight prices.

It is observed that as the number of stops increases, the flight prices also tend to increase.


The 'Total_Stops' feature is converted into numerical format using mapping, where non-stop flights are assigned 0 and increasing stops are represented by higher integers.

```python
df["Total_Stops"].unique()
```

```python
df[df["Total_Stops"].isnull()]
```

```python
df["Total_Stops"]=df["Total_Stops"].map({"non-stop":0,"1 stop":1,"2 stops":2, "3 stops":3, "4 stops":4,np.nan:0})
```
